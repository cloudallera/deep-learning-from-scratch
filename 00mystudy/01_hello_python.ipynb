- 주석 : Line 주석(# 뒤의 문자열 주석), Paragraph 주석(''' .... ''' 또는 """ ... """ 사이의 여러 줄 주석)

```python
# bool
hungry = False
sleepy = True
print(not sleepy)
print(hungry or sleepy)
```

```python
# 함수
def hello(object):       # 선언
    print("hello " + object + " !")
    
hello("cat")    # 호출
```

```python
# 클래스
class Man:       # 클래스 정의
    def __init__(self, name):
        self.name = name
        print("Initialized!!!")
        
    def hello(self):
        print("Hello " + self.name + " !")        # 인스턴스 변수 접근(self.인스턴스변수)
        
    def goodbye(self):
        print("Good-bye " + self.name + " !")
        
m = Man("David")   # 객체(인스턴스) 생성
m.hello()                 # 인스턴스 메서드 호출
m.goodbye()
```

## 1.2 numpy
- 편리하고 다양한 <u>**배열 연산**</u> 처리를 위한 패키지

```python
import numpy as np

x = np.array([1.0, 2.0, 3.0])   # python list를 numpy 배열(numpy.ndarray)로 변환
y = np.array([4.0, 5.0, 6.0])
print(x+y)
print(x / 2)     # broadcast (element wise 2로 나누기)
```

```python
A = np.array([[1, 2], [3, 4]])
A.shape
```

```python
# 형상이 다른 배열끼리 계산 (브도드캐스트 기능이 작동함)
B = np.array([10, 20])
A * B      # B가 [[10, 20], [10, 20]] 으로 확대된 후 연산됨
```

```python
X = np.array([[51, 55], [14, 19], [0, 4]])
print(X[0])
print(X[0][1])
```

```python
for row in X:
    print(row)
```

```python
X = X.flatten()
print(X)
```

```python
X[np.array([0, 2, 4])]
```

```python
X > 15
```

```python
X[X>15]    # numpy 배열의 bool 연산 결과를 numpy 배열 인덱스로 사용(index가 True 인 인자들만 선택)
```

## 1.3 matplotlib
- 그래프를 그리기 위한 시각화 라이브러리

```python
import matplotlib.pyplot as plt

# 주피터 노트북 내에 그래프를 그리기
%matplotlib inline

x = np.arange(0, 6, 0.1)       # 0에서 6까지 0.1 간격으로 생성
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin')
plt.plot(x, y2, linestyle='--', label='cos')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
# plt.show()    # ipython(jupyter notebook에서는 생략 가능)
```
